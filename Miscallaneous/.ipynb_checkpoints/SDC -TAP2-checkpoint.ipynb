```python
%%time
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np
xmin = 0
xmax = 1e6
dx = 1

tau = np.linspace(0,3,100)
y = []
for t in tau:
    x = np.arange(xmin, xmax + dx, dx)
    gx = np.pi* ((-1)**(x)) * ((2*x) + 1) * np.exp(-( ((x+0.5) * (np.pi))**2)*t)
    y.append(np.trapz(gx, x, dx)) #Caluclating summation at different tau using trapezoidal integration

#process to remove negative numbers (i.e negative flows)    
y = np.array(y)
mask = y >= 0
filtered_x = tau[mask]
filtered_y = y[mask]

# Plot the filtered data
plt.plot(filtered_x, filtered_y, label='SDC curve (Knudsen Diffusion)')
plt.xlabel(r'Dimensionless Time, $\tau$')
plt.ylabel(r'Dimensionless Exit Flow, $\bar{F_A}$')
plt.title('Standard Diffusion Curve - SDC')
plt.grid(True)
plt.show()
```
