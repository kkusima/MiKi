Note:

- For Mass to be conserved:
$$\mathbf{A}\mathbf{\nu = 0}$$

$i =$ species

$j =$ reaction

- To find the Rate of reaction of reaction j:

$$ r_j = k_{j,f} \prod_{i_f} P_{i,f} \theta_{i,f}^{A_{i,j}} - k_{j,r} \prod_{i_r} P_{i,r} \theta_{i,r}^{A_{i,j}}  $$

- To find the Rate of production of species i:
$$R_{\theta_i}=\frac{d\theta_i}{dt} = \sum_j A_{j,i} \cdot r_j $$

$$ \left[ \text{Rate of flow In} \right] - \left[ \text{Rate of flow Out} \right] + \left[\text{Rate of Generation}\right]  = \left[\text{Rate of Accumulation}\right] $$

$$ F_{A0} - F_{A} + \int_{0}^{V} r_A \,dV = \frac{dN_A}{dt} $$

Note -> Assume Steady State:

$$\frac{dN_A}{dt} = 0 $$
Therefore, 
$$F_{A0} - F_{A} + \int_{0}^{V} r_A \,dV = 0$$

As a result,
$$\frac{dF_A}{dV} = r_A $$
$$ \vdots $$
$$\frac{dF_i}{dV} = r_i $$

```python
#Reactor Info
Total_Volume = 1.0  #L
Total_Pressure = 1.0 #atm
Total_Flowrate = 658137 *(1/3600) ## Total Flowrate L/s (GHSV * Cat.Vol) * (Convertion to per s)
Volume_Step = 0.02 #Optional
Volume = np.arange(0, Total_Volume+Volume_Step, Volume_Step, dtype=float) #Volume array #Can be customized #mandatory


#Catalyst Info
##Site Densities: Pt/Pd Pd/Cu Pt/Cu Cu Only : 52 103 103 154 umol/g
Site_density = 5e-6 #moles of sites/gram of catalyst
Catalyst_gram =  0.14*(1000)*Total_Volume # mg/L -> grams
#0.14 grams of catalyst per mililitre of reactor. #GHSV
#NA = 6.022e23 #Avogadro's number (no. of molecules in a mole)

```

```python
###############Feed:
#Input Type : Mol%
Feed_Mol_pcnt = [0.000511,0 , 0.01019, 0,0.0002,0,0] #assume rest is inert if not balanced (i.e add up to 100%)

Feed_Partial_Pressures = np.array(Feed_Mol_pcnt)  * Total_Pressure
Feed_mol_frac = Feed_Partial_Pressures/Total_Pressure

F_rate = Feed_mol_frac * Total_Flowrate 
Flow_array = np.empty((len(Volume),len(F_rate)))
```

```python
init =  np.array([135,1,169,1,91,1,2000,12.467,2000,12.467,91,1,151,1,50,0.5,50,0.5,91,1])
MKM1.k = np.concatenate([init,np.ones(16)])
```

```python
%%time
#Value_tracking
ExpNoCovg = len(MKM1.Stoich.iloc[0,len(MKM1.P)+1:])
SS_covg_array = np.empty((len(Volume)+1,ExpNoCovg))
SS_covg_array[0,:] = np.append(np.zeros(ExpNoCovg-1),1)

for i in np.arange(len(Volume)-1):

    del_V = np.abs(Volume[i+1] - Volume[i])
    print(del_V)
    print(i)
    Flow_array[i,:] = F_rate
    Partial_Pressures = (Flow_array[i,:]/Total_Flowrate)*Total_Pressure
    
    MKM1.P = Partial_Pressures
    ss_covg = SS_covg_array[i,:] # MKM1.get_SS_coverages(initial_coverage=SS_covg_array[i,:])
#     SS_covg_array[i,:] = ss_covg
    
    ss_rates_prod = MKM1.get_SS_rates_production(initial_coverage=ss_covg)[:len(Feed_Partial_Pressures)]
    print(ss_rates_prod)
    ss_rates_prod = ss_rates_prod * (Site_density*Catalyst_gram)  #Conversion to give you mol/s
   
    del_F = np.multiply(ss_rates_prod, (np.ones(len(ss_rates_prod))*del_V)) #element-wise multiplication #r*delV   
    F_rate = np.add(F_rate ,del_F)
    
    SS_covg_array[i+1,:] = MKM1.get_SS_coverages(initial_coverage=ss_covg)
    ss_covg = SS_covg_array[i+1,:]
    print(ss_covg)

```
