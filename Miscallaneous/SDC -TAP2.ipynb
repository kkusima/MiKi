```python
%%time
import matplotlib.pyplot as plt
%matplotlib notebook
import numpy as np
nmin = 0
nmax = 1e5
dn = 1
n = np.arange(nmin, nmax + dn, dn)

tau = np.linspace(0.01,3.01,100)
F = []
for t in tau:
    Fn = np.pi* ((-1)**(n)) * ((2*n) + 1) * np.exp(-( ((n+0.5) * (np.pi))**2)*t)
    F.append(sum(Fn))
    
# Plot the data
plt.plot(tau, F, label='Standard Diffusion Curve')
plt.xlabel(r'Dimensionless Time, $\tau$')
plt.ylabel(r'Dimensionless Exit Flow, $\bar{F_A}$')
#plt.legend()
plt.title('Standard Diffusion Curve - SDC')
plt.grid(True)
plt.show()
```

```python
#Now plotting the concentration directly
tau = [0.005, 0.01, 0.1, 0.25, 0.5]
xi = np.linspace(0,1,100)
n = np.arange(nmin, nmax + dn, dn)

C = []
for t in tau:
    Ct = []
    for x in xi:
        Cn = 2*np.cos((n+0.5)*np.pi*x) * np.exp(-(((n+0.5)*(np.pi))**2)*t)
        Ct.append(sum(Cn))
    C.append(Ct)
   
# Plot the data
plt.figure('Concentration')
for i,t in enumerate(tau):
    plt.plot(xi, C[i], label='Concentration at '+str(t))
plt.xlabel(r'Dimensionless Reactor Coordinate, $\xi$')
plt.ylabel(r'Dimensionless Concentration, $\bar{C_A}$')
plt.legend()
plt.title('TAP Concentration Profile')
plt.grid(True)
plt.show()
```
